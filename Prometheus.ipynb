```python
import requests
from datetime import datetime, timedelta

# Define the time range you want to query
end_time = datetime.utcnow()  # End time is the current time
start_time = end_time - timedelta(days=7)  # Start time is 7 days ago

# Define the query and label config
query = 'prometheus_http_requests_total'
label_config = 'cluster="my_cluster_id",label_2="label_2_value"'

# Define the Prometheus URL and the range query endpoint
prometheus_url = 'http://10.0.101.55:30373'
query_range_endpoint = f'{prometheus_url}/api/v1/'

# Make the range query and retrieve the data
params = {
    'query': query,
    'start': start_time.timestamp(),
    'end': end_time.timestamp(),
#     'step': 15,  # Step size in seconds (default is 15 seconds)
    'label': label_config,
}
response = requests.get(query_range_endpoint, params=params)
print(response.content)
data = response.json()

# Print the data
print(data)
```
